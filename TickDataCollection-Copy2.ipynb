```python
ISU_CD_list=[]
for filename in a:
    print(filename)
    Data=pd.read_csv(os.path.join(path_dir,filename))
    ISU_CD_list +=list(Data['ISU_CD'].value_counts().index) 
    ISU_CD_list = list(set(ISU_CD_list))
    
```

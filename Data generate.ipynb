```python
def get_hlcol(Pdf):
    Pdf['현재가']=(Pdf['ASK_STEP1_BSTORD_PRC']+Pdf['BID_STEP1_BSTORD_PRC'])/2
    Pdf['High']=(Pdf['고가']-Pdf['시가'])/Pdf['시가']
    Pdf['Low']=(Pdf['저가']-Pdf['시가'])/Pdf['시가']
    Pdf['현재가']=(Pdf['현재가']-Pdf['시가'])/Pdf['시가']
    return Pdf
```

```python
def get_hmsPdf(Pdf,columnlist,outcolumnlist):
    Pdf['count']=1
    Pdf=Pdf.groupby(['ISU_CD','ORD_DD','hms']).sum()
    for col in columnlist:
        Pdf[col]=Pdf[col]/Pdf['count']
    Pdf=Pdf[outcolumnlist]
    return Pdf
```

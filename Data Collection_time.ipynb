```python
data_dir=path_dir+'/'+'ord201709'

UUID_count={}
file_list = os.listdir(data_dir)
print(file_list)
columnlist=['MBR_NO','BRN_NO']
columnISU=['ISU_CD']
filecount=0
for filename in file_list:
    print(filename)
    df1 = pd.read_csv(data_dir+'/'+filename, sep=',', header=None, encoding="cp949")
    if 'SPOT' in filename:
        df1.columns = list(layout_1.iloc[0:len(df1.iloc[0])][0])
        print("Using 2013 data")
    elif 'ksqord' in filename:

        df1.columns= header_df
        print("Using 2017 data")
    df2=df1[df1['ISU_CD'].isin(ISUlist)]
    df2 = df2[(df2['ORD_ACPT_TM']>90000000) &(df2['ORD_ACPT_TM']<153000000)]
    df2['ORD_TM']=(df2['ORD_ACPT_TM']//1e7)*36e5+ ((df2['ORD_ACPT_TM']%1e7)//1e5)*6e4+(df2['ORD_ACPT_TM']%1e5)
    df2['ORD_TM']=df2['ORD_TM']-9*36e5
    df2['time_group']=df2['ORD_TM']//(groupmin*6e4+groupsec*1e3)
    
    if(filecount==0):
        timedf=df2.groupby(['ISU_CD','ORD_DD','time_group']).last()
    else:
        timedf=pd.concat([timedf,df2.groupby(['ISU_CD','ORD_DD','time_group']).last()])
    filecount+=1
```
